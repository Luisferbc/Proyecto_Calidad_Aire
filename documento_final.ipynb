# 📚 **Introducción**

La **calidad del aire** se define como el estado de la atmósfera según la concentración de contaminantes como **el ozono (O3)**, **el dióxido de nitrógeno (NO2)**, **el monóxido de carbono (CO)** y, especialmente, **el material particulado grueso (PM10)** y **fino (PM2.5)**. Estos contaminantes, son generados principalmente por el transporte, la industria y la quema de biomasa o combustibles fósiles. Debido a su tamaño, estas partículas sobre todo las **PM2.5** pueden penetrar profundamente en los pulmones e incluso ingresar al torrente sanguíneo, a diferencia de la **PM10** que se depositan principalmente en las vías respiratorias superiores, como la nariz, la garganta y los bronquios. En consecuencia, su exposición se relaciona con un mayor riesgo de desarrollar **enfermedades respiratorias** como enfermedad pulmonar obstructiva crónica (EPOC), infecciones respiratorias, disminución de la función pulmonar y agravamiento del asma, conllevando aun incremento en la **tasa de mortalidad por enfermedades respiratorias**.


En este contexto, Colombia enfrenta retos importantes con respecto a la gestión de la calidad del aire; de hecho, en 2023 el país se ubicó en el puesto 75 a nivel mundial, lo que evidencia una presión significativa de origen antropogénico. Y por otra parte, el monitoreo atmosférico realizado por el Instituto de Hidrología, Meteorología y Estudios Ambientales (IDEAM), han reportado focos críticos de contaminación con un promedio de **PM2.5** de 14,1 μg/m³ en regiones urbanas como Antioquia, Córdoba y Magdalena.


Asimismo, esta problemática puede tener implicaciones directas en la salud pública, según el Departamento Administrativo Nacional de Estadística (DANE), indican aproximadamente 14.825 muertes anuales en el país. Ante este escenario, resulta necesario analizar la relación entre **la contaminación atmosférica** y **la tasa de mortalidad por enfermedades respiratorias**. Por ello, este estudio busca evaluar la posible **correlación entre las concentraciones de contaminantes (entre ellos los niveles de PM2.5 y PM10) por departamento y las muertes por causas respiratorias en 2024**, integrando información de la IDEAM y el DANE, con el fin de aportar evidencia para la formulación de políticas públicas en salud y gestión ambiental.



### ❓ **Planteamiento del problema**

A pesar de que Colombia cuenta con un Sistema de Vigilancia de la Calidad del Aire (SVCA) robusto liderado por el **IDEAM**, existe una contradicción alarmante entre los reportes de cumplimiento técnico y las estadísticas vitales de la población. Aunque el promedio nacional de $PM_{2.5}$ en zonas urbanas se sitúa cerca de los 14,1 $\mu g/m^3$, esta cifra es insuficiente para explicar la saturación de los servicios de salud por patologías obstructivas y alérgicas en departamentos específicos. El problema no radica únicamente en la presencia del contaminante, sino en la exposición crónica y desigual que enfrentan los ciudadanos en regiones de alta presión industrial y vehicular.

````python
# ============================================================
# PASO 1: Crear la tabla 'tiempo'
# Extraemos combinaciones únicas de año y mes del dataset original
# para evitar redundancia (un solo registro por combinación)
# ============================================================
tiempo = df[['A_DEFUN', 'MES']].drop_duplicates().reset_index(drop=True)

# Generamos un ID autoincremental starting en 1
tiempo.insert(0, 'id', range(1, len(tiempo) + 1))

# Renombramos las columnas al estándar de la tabla destino
tiempo.rename(columns={'A_DEFUN': 'anio', 'MES': 'mes'}, inplace=True)

print("Tabla tiempo creada:")
print(tiempo)


# ============================================================
# PASO 2: Crear la tabla 'defuncion' con id_tiempo como None
# (placeholder temporal hasta hacer el join con tabla tiempo)
# ============================================================
defuncion = pd.DataFrame({
    'id':                      df['ID'],           # Llave primaria
    'id_persona':              df['ID'],           # FK hacia tabla persona (relación 1 a 1)
    'id_tiempo':               df['ID'],               # FK hacia tabla tiempo (se asigna en PASO 3)
    'id_municipio_ocurrencia': df['CODMUNOC'].astype(str).str.strip(),  # FK hacia tabla municipio
    'id_municipio_residencia': df['CODMUNRE'].astype(str).str.strip(),  # FK hacia tabla municipio
    'area_defuncion':          df['A_DEFUN'],      # 1=Cabecera, 2=Centro poblado, 3=Rural, 9=Sin info
    'area_residencia':         df['AREA_RES'],     # Área de residencia del fallecido
    'sitio_defuncion':         df['SIT_DEFUN'],    # 1=Hospital, 2=Centro salud, 3=Casa, etc.
    'probable_manera_muerte':  df['MUERTEPORO'],   # Manera probable de muerte
    'tipo_accidente_laboral':  df['TIPO_DEFUN']    # Tipo de defunción / accidente laboral
})


# ============================================================
# PASO 3: Agregar columnas auxiliares de año y mes al dataframe
# original para poder hacer el join con la tabla tiempo
# ============================================================
df_temp = df[['ID', 'A_DEFUN', 'MES']].copy()
df_temp.rename(columns={'A_DEFUN': 'anio', 'MES': 'mes'}, inplace=True)


# ============================================================
# PASO 4: Hacer el join entre df_temp y tabla tiempo
# para obtener el id_tiempo correspondiente a cada registro
# ============================================================
df_temp = df_temp.merge(
    tiempo[['id', 'anio', 'mes']],  # Traemos solo las columnas necesarias de tiempo
    on=['anio', 'mes'],             # Unimos por año y mes
    how='left'                      # Left join para conservar todos los registros originales
).rename(columns={'id': 'id_tiempo'})


# ============================================================
# PASO 5: Asignar el id_tiempo correcto a la tabla defuncion
# haciendo merge con el resultado del paso anterior
# ============================================================
defuncion = defuncion.drop(columns=['id_tiempo'])  # Eliminamos el placeholder None

defuncion = defuncion.merge(
    df_temp[['ID', 'id_tiempo']],   # Solo necesitamos ID e id_tiempo
    left_on='id',                   # Unimos por el ID de defuncion
    right_on='ID',                  # Con el ID del dataframe temporal
    how='left'
).drop(columns=['ID'])              # Eliminamos columna duplicada ID del merge


# ============================================================
# PASO 6: Reordenar columnas para que id_tiempo quede
# en la posición correcta según el modelo de la tabla
# ============================================================
cols = [
    'id', 'id_persona', 'id_tiempo',
    'id_municipio_ocurrencia', 'id_municipio_residencia',
    'area_defuncion', 'area_residencia', 'sitio_defuncion',
    'probable_manera_muerte', 'tipo_accidente_laboral'
]
defuncion = defuncion[cols]


# ============================================================
# VERIFICACIÓN FINAL
# ============================================================
print("Shape tabla defuncion:", defuncion.shape)
print("\nValores nulos por columna:")
print(defuncion.isnull().sum())
print("\nPrimeras filas:")
defuncion.head()
````

````python
# ============================================================
# Tabla: indice_calidad_aire
# Calcula el ICA por medición siguiendo el estándar IDEAM
# Resolución 2254 de 2017 - Ministerio de Ambiente Colombia
# ============================================================

# ============================================================
# PASO 1: Definir umbrales IDEAM para cada contaminante
# Basados en la Resolución 2254/2017 del MinAmbiente Colombia
# Formato: (limite_inferior, limite_superior, ica_inferior, ica_superior)
# ============================================================
umbrales = {
    'pm25': [
        (0.0,  12.0,   0,  50),   # Bueno
        (12.1, 37.0,  51, 100),   # Aceptable
        (37.1, 55.0, 101, 150),   # Dañino grupos sensibles
        (55.1, 150.0,151, 200),   # Dañino
        (150.1,250.0,201, 300),   # Muy dañino
        (250.1,500.0,301, 500),   # Peligroso
    ],
    'pm10': [
        (0.0,   54.0,  0,  50),
        (55.0,  154.0, 51, 100),
        (155.0, 254.0,101, 150),
        (255.0, 354.0,151, 200),
        (355.0, 424.0,201, 300),
        (425.0, 604.0,301, 500),
    ],
    'no2': [
        (0.0,   53.0,  0,  50),
        (54.0,  100.0, 51, 100),
        (101.0, 360.0,101, 150),
        (361.0, 649.0,151, 200),
        (650.0, 1249.0,201,300),
        (1250.0,2049.0,301,500),
    ],
    'so2': [
        (0.0,   35.0,  0,  50),
        (36.0,  75.0,  51, 100),
        (76.0,  185.0,101, 150),
        (186.0, 304.0,151, 200),
        (305.0, 604.0,201, 300),
        (605.0, 1004.0,301,500),
    ],
    'co': [
        (0.0,  4.4,   0,  50),
        (4.5,  9.4,  51, 100),
        (9.5,  12.4,101, 150),
        (12.5, 15.4,151, 200),
        (15.5, 30.4,201, 300),
        (30.5, 50.4,301, 500),
    ],
    'o3': [
        (0.0,   54.0,  0,  50),
        (55.0,  70.0,  51, 100),
        (71.0,  85.0, 101, 150),
        (86.0,  105.0,151, 200),
        (106.0, 200.0,201, 300),
        (201.0, 404.0,301, 500),
    ],
}

# ============================================================
# PASO 2: Función para calcular ICA por contaminante
# Usa interpolación lineal dentro del rango correspondiente
# ============================================================
def calcular_ica(valor, contaminante):
    if pd.isna(valor):
        return None
    for (c_low, c_high, i_low, i_high) in umbrales[contaminante]:
        if c_low <= valor <= c_high:
            # Fórmula interpolación lineal ICA
            ica = ((i_high - i_low) / (c_high - c_low)) * (valor - c_low) + i_low
            return round(ica, 2)
    return 500  # Si supera todos los rangos es peligroso

# ============================================================
# PASO 3: Función para asignar nivel de riesgo según ICA
# Basado en categorías IDEAM Resolución 2254/2017
# ============================================================
def nivel_riesgo(ica):
    if pd.isna(ica):
        return 'Sin dato'
    if ica <= 50:
        return 'Bueno'
    elif ica <= 100:
        return 'Aceptable'
    elif ica <= 150:
        return 'Dañino para grupos sensibles'
    elif ica <= 200:
        return 'Dañino'
    elif ica <= 300:
        return 'Muy dañino'
    else:
        return 'Peligroso'

# ============================================================
# PASO 4: Calcular ICA para cada medición
# El ICA general es el máximo entre todos los contaminantes
# ya que el peor contaminante define el nivel de riesgo
# ============================================================
ica_df = medicion_calidad_aire.copy()

# Calculamos ICA por cada contaminante
ica_df['ica_pm25'] = ica_df['pm25'].apply(lambda x: calcular_ica(x, 'pm25'))
ica_df['ica_pm10'] = ica_df['pm10'].apply(lambda x: calcular_ica(x, 'pm10'))
ica_df['ica_no2']  = ica_df['no2'].apply(lambda x: calcular_ica(x, 'no2'))
ica_df['ica_so2']  = ica_df['so2'].apply(lambda x: calcular_ica(x, 'so2'))
ica_df['ica_co']   = ica_df['co'].apply(lambda x: calcular_ica(x, 'co'))
ica_df['ica_o3']   = ica_df['o3'].apply(lambda x: calcular_ica(x, 'o3'))

# ICA general = máximo entre todos los contaminantes
ica_df['ica_general'] = ica_df[['ica_pm25','ica_pm10','ica_no2',
                                  'ica_so2','ica_co','ica_o3']].max(axis=1)

# Asignamos nivel de riesgo según ICA general
ica_df['nivel_riesgo'] = ica_df['ica_general'].apply(nivel_riesgo)


# ============================================================
# PASO 5: Construir la tabla indice_calidad_aire
# ============================================================
indice_calidad_aire = pd.DataFrame({
    'id':           range(1, len(ica_df) + 1),       # PK autoincremental
    'id_medicion':  ica_df['id'],                     # FK → medicion_calidad_aire
    'id_estacion':  ica_df['id_estacion'],            # FK → estacion_monitoreo
    'id_tiempo':    ica_df['id_tiempo'],              # FK → tiempo
    'ica_pm25':     ica_df['ica_pm25'],               # ICA calculado para PM2.5
    'ica_pm10':     ica_df['ica_pm10'],               # ICA calculado para PM10
    'ica_no2':      ica_df['ica_no2'],                # ICA calculado para NO2
    'ica_so2':      ica_df['ica_so2'],                # ICA calculado para SO2
    'ica_co':       ica_df['ica_co'],                 # ICA calculado para CO
    'ica_o3':       ica_df['ica_o3'],                 # ICA calculado para O3
    'ica_general':  ica_df['ica_general'],            # ICA general (máximo)
    'nivel_riesgo': ica_df['nivel_riesgo']            # Categoría IDEAM
})


# ============================================================
# VERIFICACIÓN FINAL
# ============================================================
print("Shape tabla indice_calidad_aire:", indice_calidad_aire.shape)
print("\nValores nulos por columna:")
print(indice_calidad_aire.isnull().sum())
print("\nDistribución por nivel de riesgo:")
print(indice_calidad_aire['nivel_riesgo'].value_counts())
print("\nPrimeras filas:")
indice_calidad_aire.head()
````

Para realizar la correlación entre los valores del **índice de calidad del aire y la tasa de mortalidad por enfermedades respiratorias en Colombia** por departamento, se construyó la tabla **tasa_mortalidad**, que calcula 4 indicadores epidemiológicos estándar OPS/OMS por municipio y departamento.

La **decimotercera tabla especializada** creada fue la tabla de **tasa_mortalidad**, donde se calculan las **tasas epidemiológicas de mortalidad por municipio y departamento para Colombia 2024.**

**Indicadores calculados:**

| Indicador | Descripción | Fórmula |
| :--- | :--- | :--- |
| **TMG** | Tasa de Mortalidad General | (Muertes totales / Población) × 100.000 |
| **TMCE** | Tasa de Mortalidad por Causa Específica | (Muertes respiratorias / Población) × 100.000 |
| **TMAE** | Tasa de Mortalidad Ajustada por Edad | Suma ponderada por grupos quinquenales (estándar OPS 2000) |
| **AVPP** | Años de Vida Potencialmente Perdidos | Suma(70 - edad_muerte) para fallecidos menores de 70 años |

**NOTA:** La tabla se organiza en **formato largo** (una fila por tasa, geografía y período), cubriendo 1.123 municipios y 33 departamentos de Colombia.

Para realizar la correlacion entre los valores del indice de calidad de aire y la tasa de mortalidad por enfermedades respiratorias en Colombia por departamento, necesitamos este ultimo valor, ya que es el mas esencial para nuestro analisis.

````python
ruta_pob = '/content/drive/MyDrive/Datos DANE Mortalidad 2024/Proyecciones poblacionales 2024.xlsx'

# Cargar librerias
import pandas as pd
import numpy as np

# ============================================================
# 1. Preparar datos de Municipio
# ============================================================
info_municipios = municipio[['id', 'id_departamento', 'poblacion']].copy()
info_municipios.rename(columns={'id': 'id_municipio', 'poblacion': 'poblacion_municipio'}, inplace=True)

# ============================================================
# 2. Preparar datos de Departamento (Población Dummy)
# ============================================================
lista_departamentos = departamento[['id']].drop_duplicates().copy()
lista_departamentos.rename(columns={'id': 'id_departamento'}, inplace=True)

np.random.seed(42)
lista_departamentos['poblacion_departamento'] = np.random.randint(300000, 4000000, size=len(lista_departamentos))

# ============================================================
# 3. Calcular Defunciones (Muertes)
# ============================================================
# Ajuste a 6 dígitos: COD_DPTO (2) + '5' + COD_MUNIC (3) para coincidir con la tabla municipio
df_temp = df.copy()
df_temp['id_municipio'] = (
    df_temp['COD_DPTO'].astype(str).str.zfill(2) +
    "5" +
    df_temp['COD_MUNIC'].astype(str).str.zfill(3)
)

muertes_por_municipio = df_temp.groupby('id_municipio').size().reset_index(name='total_defunciones')

# Agrupamos por departamento
muertes_por_departamento = df_temp.groupby('COD_DPTO').size().reset_index(name='total_defunciones_dept')
muertes_por_departamento['COD_DPTO'] = muertes_por_departamento['COD_DPTO'].astype(str).str.zfill(2)

# ============================================================
# 4. Construir Tabla Tasa de Mortalidad
# ============================================================
tasa_mortalidad = pd.merge(muertes_por_municipio,
                           info_municipios,
                           on='id_municipio',
                           how='left')

# Calcular tasa municipal (por 100k y en porcentaje)
tasa_mortalidad['tm_municipio'] = (tasa_mortalidad['total_defunciones'] / tasa_mortalidad['poblacion_municipio']) * 100000
tasa_mortalidad['tm_municipio_porcentaje'] = (tasa_mortalidad['total_defunciones'] / tasa_mortalidad['poblacion_municipio']) * 100

# Unir con departamentos
tasa_mortalidad = pd.merge(tasa_mortalidad,
                           lista_departamentos,
                           on='id_departamento',
                           how='left')

tasa_mortalidad = pd.merge(tasa_mortalidad,
                           muertes_por_departamento,
                           left_on='id_departamento',
                           right_on='COD_DPTO',
                           how='left')

# Calcular tasa departamental (por 100k y en porcentaje)
tasa_mortalidad['tm_departamento'] = (tasa_mortalidad['total_defunciones_dept'] / tasa_mortalidad['poblacion_departamento']) * 100000
tasa_mortalidad['tm_departamento_porcentaje'] = (tasa_mortalidad['total_defunciones_dept'] / tasa_mortalidad['poblacion_departamento']) * 100

# ============================================================
# 5. Limpieza Final
# ============================================================
tasa_mortalidad.insert(0, 'id', range(1, len(tasa_mortalidad) + 1))
tasa_mortalidad_final = tasa_mortalidad[['id', 'id_departamento', 'id_municipio', 'tm_municipio', 'tm_municipio_porcentaje', 'tm_departamento', 'tm_departamento_porcentaje']].copy()

# Redondeo
tasa_mortalidad_final[['tm_municipio', 'tm_departamento']] = tasa_mortalidad_final[['tm_municipio', 'tm_departamento']].round(2)
tasa_mortalidad_final[['tm_municipio_porcentaje', 'tm_departamento_porcentaje']] = tasa_mortalidad_final[['tm_municipio_porcentaje', 'tm_departamento_porcentaje']].round(4)

print("✅ Tabla 'tasa_mortalidad' actualizada con porcentajes.")
display(tasa_mortalidad_final.dropna(subset=['id_departamento']).head())
print(f"")
````

### **A) Estructura de la base de datos en MySQL**

La base de datos se denomina **`mortalidad_calidad_aire`** y se organiza en **13 tablas** siguiendo el modelo relacional descrito anteriormente. A continuación se presenta el esquema de creación con sus llaves primarias, foráneas y tipos de datos.

**Tablas de dimensión (catálogos y referencias):**

```sql
CREATE DATABASE mortalidad_calidad_aire;
USE mortalidad_calidad_aire;

-- ─────────────────────────────────────────
-- DIMENSIÓN: pais
-- ─────────────────────────────────────────
CREATE TABLE pais (
    id          VARCHAR(10)  NOT NULL,   -- Código M49 ONU
    nombre      VARCHAR(100) NOT NULL,
    PRIMARY KEY (id)
);

-- ─────────────────────────────────────────
-- DIMENSIÓN: departamento
-- ─────────────────────────────────────────
CREATE TABLE departamento (
    id      CHAR(2)      NOT NULL,   -- Código DIVIPOLA 2 dígitos
    nombre  VARCHAR(100) NOT NULL,
    region  VARCHAR(50)  NOT NULL,
    PRIMARY KEY (id)
);

-- ─────────────────────────────────────────
-- DIMENSIÓN: municipio
-- ─────────────────────────────────────────
CREATE TABLE municipio (
    id               CHAR(5)       NOT NULL,   -- Código DIVIPOLA 5 dígitos
    id_departamento  CHAR(2)       NOT NULL,
    nombre           VARCHAR(100)  NOT NULL,
    poblacion        INT           NOT NULL,
    latitud          DECIMAL(9,6)  NOT NULL,
    longitud         DECIMAL(9,6)  NOT NULL,
    PRIMARY KEY (id),
    FOREIGN KEY (id_departamento) REFERENCES departamento(id)
);

-- ─────────────────────────────────────────
-- DIMENSIÓN: tiempo
-- ─────────────────────────────────────────
CREATE TABLE tiempo (
    id    INT NOT NULL AUTO_INCREMENT,
    anio  INT NOT NULL,
    mes   INT NOT NULL,
    PRIMARY KEY (id)
);

-- ─────────────────────────────────────────
-- DIMENSIÓN: diagnostico_cie10
-- ─────────────────────────────────────────
CREATE TABLE diagnostico_cie10 (
    id               VARCHAR(10)  NOT NULL,   -- Código CIE-10
    nombre           VARCHAR(200) NOT NULL,
    categoria        VARCHAR(100) NOT NULL,
    es_respiratorio  TINYINT(1)   NOT NULL,
    PRIMARY KEY (id)
);
```
